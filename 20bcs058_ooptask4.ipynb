```python
# problem 1
class Node:
    # Class to create nodes of linked list
    # constructor initializes node automatically
    def __init__(self, data=None):
        self.data = data
        self.next = None

class stack:
    def __init__(self): 
        self.head=None  # head is default NULL
   
    # Checks if stack is empty
    def isempty(self):
        if self.head==None:
            return True
        else:
            return False     

    # Method to add data to the stack
    # adds to the start of the stack
    def push(self,data):

        if self.head==None:
            self.head=Node(data)
        else:
            newnode=Node(data)
            newnode.next=self.head
            self.head=newnode
    
    # Remove element that is the current head (start of the stack)            
    def pop(self):
        if self.isempty():
            return None

        else:
       # Removes the head node and makes
       #the preceeding one the new head    
             poppednode=self.head
             self.head=self.head.next
             poppednode.next=None
             return poppednode.data

     # Returns the head node data
    def peek(self):
         
        if self.isempty():
            return None
             
        else:
            return self.head.data
             
     # Prints out the stack    
    def display(self):
         
        iternode = self.head
        if self.isempty():
            print("Stack Underflow")
         
        else:
             
            while(iternode.next != None):
                 
                print(iternode.data,"->",end = " ")
                iternode = iternode.next
            print(iternode.data)
    def __del__(self):
        print("die")         
mystack=stack()
mystack.push(10)
mystack.push(20)
mystack.push(30)
mystack.push(40)

mystack.display()
print("Top element is ",mystack.peek())
mystack.pop()
mystack.pop()

mystack.display()
print("Top element is ",mystack.peek())
```

```python
# problem 3         
class Node:
     
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
     
    def __init__(self):
        self.front = self.rear = None
 
    def isEmpty(self):
        return self.front == None
  # Method to add an item to the queue
    def EnQueue(self, item):
        temp = Node(item)
         
        if self.rear == None:
            self.front = self.rear = temp
            return
        self.rear.next = temp
        self.rear = temp
 
    # Method to remove an item from queue
    def DeQueue(self):
         
        if self.isEmpty():
            return
        temp = self.front
        self.front = temp.next
 
        if(self.front == None):
            self.rear = None
       
    def __del__(self):
        print("die")    
q = Queue()
q.EnQueue(10)
q.EnQueue(20)
q.DeQueue()
q.DeQueue()
q.EnQueue(30)
q.EnQueue(40)
q.EnQueue(50)
q.DeQueue()  
print("Queue Front " + str(q.front.data))
print("Queue Rear " + str(q.rear.data))
               
```
